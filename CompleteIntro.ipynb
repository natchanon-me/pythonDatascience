For loop

```
# Loop in defined list
for i in [1,2,3]:
  print(i)
print("################")
# Print character in string
for item in 'Test':
  print(item)
print("################")
# Loop for any range ------> range arguments : range(start, end, increment)
for i in range(2):
  print(i)
print("################")
for i in range(1, 15, 3):
  print(i)
print("################")
```

***Pandas Series***

```
# function -->  Series in pandas is dict in python 1D array
import pandas as pd

dict1 = {'A' : 10, 'B' : 20, 'C' : 30}
print(dict1)

### PANDAS ####

series = pd.Series(dict1)
print(series)
print(type(series))

print(series.index)
print(series.values)
```

```
names = ["few","da","fern","tu","pu"]
heights = [176, 156, 162, 158, 168]

pandas = pd.Series(heights, names)

# Accessing data can be proceed by key(label) or index
print(pandas["few"])
print(pandas[0])
print(pandas[["da", "fern", "pu"]])
print(pandas[[0, 2, 4]])
print(pandas[[True,False,True,False,True]]) # Can access by boolean operation
```

***Pandas Dataframe***

```
# Array 2D (Matrix)
# Dataframe
values = np.random.randint(0, 20, (3,3))
rows = ['A', 'B', 'C']
columns = ['X', 'Y', 'Z']

df = pd.DataFrame(values, rows, columns)
print(df)

#Access
print(df['X'][['A', 'B']])
print(df.loc['A']) # Accessing by row label
print(df.iloc[1]) # Accessing by row index
```

### ***MATPLOTLIB***


```
# Example code
import numpy as np
import matplotlib.pyplot as plt


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.2       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
